#Model Development

```python
df=df.drop('total sulfur dioxide', axis=1)

#the total sulphur dioxide and free sulphur dioxide are highly correlated features so, I remove one of them
```

```python
#Replacing dummy variables
df['type']=pd.get_dummies(df['type'],drop_first=True)

#OR: df.replace({'white': 1, 'red': 0}, inplace=True)
```

```python
df['best quality']=[1 if x>5 else 0 for x in df.quality]
```

```python
X=df.drop(['quality','best quality'], axis=1)
y=df['best quality']
```

```python
#Split our data to train and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=40)
```

```python
X_train.shape, X_test.shape
```

```python
#Normalizing data
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#Training machine learning models
lr_model=LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
lr_prediction=lr_model.predict(X_test)
```

```python
xgbc_model=XGBClassifier()
```

```python
xgbc_model.fit(X_train,y_train)
xgbc_prediction=xgbc_model.predict(X_test)
```

```python
svc_model=SVC(kernel='rbf')
```

```python
svc_model.fit(X_train,y_train)
svc_prediction=svc_model.predict(X_test)
```

#Model Evaluation

```python
print('LogisticRegression:')
print(metrics.roc_auc_score(y_test,lr_prediction))
print('XGBClassifier:')
print(metrics.roc_auc_score(y_test,xgbc_prediction))
print('SVC:')
print(metrics.roc_auc_score(y_test,svc_prediction))
```

```python
print('LogisticRegression:')
print(metrics.confusion_matrix(y_test,lr_prediction))
print('XGBClassifier:')
print(metrics.confusion_matrix(y_test,xgbc_prediction))
print('SVC:')
print(metrics.confusion_matrix(y_test,svc_prediction))
```

```python
print('LogisticRegression:')
print(metrics.classification_report(y_test,lr_prediction))
print('XGBClassifier:')
print(metrics.classification_report(y_test,xgbc_prediction))
print('SVC:')
print(metrics.classification_report(y_test,svc_prediction))
```
